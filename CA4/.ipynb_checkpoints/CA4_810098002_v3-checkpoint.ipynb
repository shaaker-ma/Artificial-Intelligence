### Numerizing "brand"

```python
def brandCategorizer(data, brand, categories):
    for i in range(len(data[brand])):
        brand_i = data[brand][i]
        if(brand_i in categories):
            data[brand][i] = categories.get(brand_i)
    return data
```

```python
uniqueCategories = data['brand'].unique()
categories ={}
for i in range(len(uniqueCategories)):
    categories[uniqueCategories[i]] = i
for category in categories:
    print("Mapped: ",category," -> ", categories.get(category))
data = brandCategorizer(data, 'brand', categories)
data
```

```python
def dataCleanse(df,col):
    normalizer = Normalizer()
    lemmatizer = Lemmatizer() 
    
    for i in range(len(df[col])):
        df[col][i] = normalizer.normalize(df[col][i])
            
    punctuations = set('''!()-[]{};:'"\,٬٫،<>./?@#$%٪^&*×ـ+=_~.؟؛«»|…''') #Set of unnecessary punctuations
    for i in range(len(df[col])):
        without_punctuations = ""
        for char in df[col][i]:
            if(char not in punctuations):
                without_punctuations = without_punctuations + char
        df[col][i] = word_tokenize(without_punctuations)
        
    stops = set(stopwords_list()) #List of unnecessary words
    for i in range(len(df[col])):
        wordsWithoutStops = ""
        for j in range(len(df[col][i])):
            if(df[col][i][j] not in stops):
                wordsWithoutStops = wordsWithoutStops + " " + (lemmatizer.lemmatize(df[col][i][j]))
        df[col][i] = wordsWithoutStops
    return df
```

# Question 2 answer: Categorical Feature

In this question I will answer why do we need numerical feature instead of categorical feature.
In machine learniong we work with numbers not texts, So in order to perform machine learning algorithms, we need to change categorical to numerical features.
There are ways we can numeric our categorical features:

- Approach 1: Label Encoding
        Our approach to encoding categorical values is to use a technique called label encoding. Label encoding is simply converting each value in a column to a number. For example, the body_style column contains 5 different values. We could choose to encode it like this:
        Mapped:  Wednesday 04AM  ->  0
        Mapped:  Wednesday 11AM  ->  1
        Mapped:  Wednesday 01PM  ->  2
        Mapped:  Thursday 07AM  ->  3
        Mapped:  Thursday 08AM  ->  4
        Mapped:  Thursday 09AM  ->  5
        Mapped:  Friday 04AM  ->  6
        Mapped:  Friday 12PM  ->  7
        Mapped:  Saturday 06AM  ->  8
        Mapped:  Saturday 12PM  ->  9
        Mapped:  Saturday 01PM  ->  10
        Mapped:  Sunday 08AM  ->  11
        Mapped:  Sunday 09AM  ->  12
        Mapped:  Sunday 11AM  ->  13
        ...
- Approach 2: One Hot Encoding
        Label encoding has the advantage that it is straightforward but it has the disadvantage that the numeric values can be “misinterpreted” by the algorithms. For example, the value of 0 is obviously less than the value of 4 but does that really correspond to the data set in real life? Does a wagon have “4X” more weight in our calculation than the convertible? In this example, I don’t think so.

        A common alternative approach is called one hot encoding (but also goes by several different names shown below). Despite the different names, the basic strategy is to convert each category value into a new column and assigns a 1 or 0 (True/False) value to the column. This has the benefit of not weighting a value improperly but does have the downside of adding more columns to the data set.

# Question 3 answer:  Feature extraction from text using CountVectorizer & Tf-idf Vectorizer

Vectorization is the general process of turning a collection of text documents into numerical feature vectors. This specific strategy (tokenization, counting and normalization) is called the Bag of Words or “Bag of n-grams” representation. Documents are described by word occurrences while completely ignoring the relative position information of the words in the document.

We can see that we cannot work with this data for our processing, so we will have to change this feature to a numerical feature, we can use Count vectorizer or Tf-idf vectorizer.





- Example of TF-IDF Vectorizer:

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfTransformer as TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
dataset = [
    "I enjoy reading about Machine Learning and Machine Learning is my PhD subject",
    "I would enjoy a walk in the park",
    "I was reading in the library"]
tfIdfTransformer = TfidfTransformer(use_idf=True)
countVectorizer = CountVectorizer()
wordCount = countVectorizer.fit_transform(dataset)
newTfIdf = tfIdfTransformer.fit_transform(wordCount)
df = pd.DataFrame(newTfIdf[0].T.todense(), index=countVectorizer.get_feature_names(), columns=["TF-IDF"])
df = df.sort_values('TF-IDF', ascending=False)
print (df.head(10))
```

- Example of Count Vectorizer:

Here is 5 solutions for missing data:
    
    1- Do Nothing:
        well this is the most easy thing there is to do, and obviously we are not going to use it compare to the average
        replacement
        
    2- Imputation Using (Mean/Median) Values:
        Our method above.
        
        -Disadvantages:
            -Mean imputation does not preserve the relationships among variables.
            -Mean Imputation Leads to An Underestimate of Standard Errors
            
        -Advantages:
            -Super easy
    
    3- Imputation Using Most Frequent or Zero/Constant Values: 
        Most Frequent is another statistical strategy to impute missing values 
        
       - Benifits:
            -Works well with categorical features.
            
       - Consequences:
            -It does NOT factor the correlations between features
            - can introduce bias in the data.
    
    4-Imputation Using k-NN: -> Better than 3 above
        KNN or k nearest neighbours is used for simple classification, it uses feature similarity to predict the values
        of any new data points.This means that the new point is assigned a value based on how closely it resembles the points
        in the training set.
        
        - Benifits:
            -Much more accurate than the mean, median or most frequent imputation methods.
    
    5- Deleting those rows
        It gets benefitial when our dataset is large and deleting some rows wouldn't hit our dataset in a bad way.

```python
def rmse(predictions, targets):
    Sum = 0
    for i in range(len(predictions)):
        Sum = Sum + (predictions[i] - targets[i])**2
    Sum = Sum/(len(predictions))
    res = sqrt(Sum)
    return res
```

```python
def mse(predictions, targets):
    Sum = 0
    for i in range(len(predictions)):
        Sum = Sum + (predictions[i] - targets[i])**2
    Sum = Sum/(len(predictions))
    res = Sum
    return res
```

## KNN
In statistics, the k-nearest neighbors algorithm (k-NN) is a non-parametric classification method first developed by Evelyn Fix and Joseph Hodges in 1951 and later expanded by Thomas Cover. It is used for classification and regression. In both cases, the input consists of the k closest training examples in data set. The output depends on whether k-NN is used for classification or regression

### Question: n_neighber:  K to your chosen number of neighbors, the input consists of the k closest training examples in data set.

# Question 5: Data division

I have devided my data into 80% train data and 20% test data, and it is because in a lot of machine learning algorithms and even in neural networks, this is what we do, we take 80% of data as our training data and for that 80%, we will take 80% as pure training and 20% for our validation test. and it is most of the time really benefitial.

## 98% train data and 20% test -> usually overfits

Overfitting is a modeling error in statistics that occurs when a function is too closely aligned to a limited set of data points. As a result, the model is useful in reference only to its initial data set, and not to any other data sets.

I have brought an example of overfitting on data from my Deep-Learning course this semester, it shows that the more we get biased on our training set, the more distant we will become from validation



## 40% train data and 60% test -> usually underfits

A statistical model or a machine learning algorithm is said to have underfitting when it cannot capture the underlying trend of the data. Underfitting destroys the accuracy of our machine learning model. Its occurrence simply means that our model or the algorithm does not fit the data well enough. It usually happens when we have less data to build an accurate model and also when we try to build a linear model with a non-linear data. In such cases the rules of the machine learning model are too easy and flexible to be applied on such minimal data and therefore the model will probably make a lot of wrong predictions. Underfitting can be avoided by using more data and also reducing the features by feature selection.

# Decision Tree 

Decision Trees (DTs) are a non-parametric supervised learning method used for classification and regression. The goal is to create a model that predicts the value of a target variable by learning simple decision rules inferred from the data features. A tree can be seen as a piecewise constant approximation.

Decision-tree learners can create over-complex trees that do not generalise the data well. This is called overfitting. Mechanisms such as pruning, setting the minimum number of samples required at a leaf node or setting the maximum depth of the tree are necessary to avoid this problem.

### Question: maximum depth : 
    the number of samples required to populate the tree doubles for each additional level the tree grows to.
Effect:


### Question: min_samples_split: 
    We use min_samples_split or min_samples_leaf to ensure that multiple samples inform every decision in the tree, by controlling which splits will be considered. 
    A very small number will usually mean the tree will overfit, whereas a large number will prevent the tree from learning the data. Try min_samples_leaf=5 as an initial value. If the sample size varies greatly, a float number can be used as percentage in these two parameters.


# Question 6:

#### Question: If max_Depth would be too large: 
    If you set max_depth too high, then the decision tree might simply overfit the training data without capturing useful patterns as we would like; this will cause testing error to increase.

#### Question: If max_Depth would be too large: 
    If we set it too low, that is not good as well; then you might be giving the decision tree too little flexibility to capture the patterns and interactions in the training data. This will also cause the testing error to increase.


```python
max_Depths = [1, 5, 10, 30, 60, 80, 100, 200]
res_mse_list = []
res_rmse_list = []
for depth in max_Depths:
    dtree = DecisionTreeRegressor(max_depth = depth)
    dtree = dtree.fit(x_train, y_train)
    predition = dtree.predict(x_test)
    predition_list = predition.tolist()
    y_test_list = y_test.tolist()
    res_mse_list.append(mse(predition_list, y_test_list))
    res_rmse_list.append(rmse(predition_list, y_test_list))
```

```python
plt.figure(figsize = (10,7))
plt.plot(max_Depths, res_mse_list, color='b', label='MSE for max_depths')
plt.xlabel("Max Depths")
plt.ylabel("MSE")
plt.title("MSE for max_depths")
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10,7))
plt.plot(max_Depths, res_rmse_list, color='r', label='RMSE for max_depths')
plt.xlabel("Max Depths")
plt.ylabel("RMSE")
plt.title("RMSE for max_depths")
plt.legend()
plt.show()
```

We can totally see that everything that has been stated so far is working properly and 

- If you set max_depth too high, then the decision tree might simply overfit the training data without capturing useful patterns as we would like; this will cause testing error to increase.

- If we set it too low, that is not good as well; then you might be giving the decision tree too little flexibility to capture the patterns and interactions in the training data. This will also cause the testing error to increase.

# Random Forest

A random forest is a meta estimator that fits a number of classifying decision trees on various sub-samples of the dataset and uses averaging to improve the predictive accuracy and control over-fitting. The sub-sample size is controlled with the max_samples parameter if bootstrap=True (default), otherwise the whole dataset is used to build each tree.

#### Question: max_depth
    The maximum depth of the tree. If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples.
    
#### Question  n_estimators
    The number of trees in the forest. By default it is 100
    

```python
max_Depths = [10, 40, 80, 100]
n_estimators = [20, 60, 100, 160]
res_mse_list = []
res_rmse_list = []
for depth in max_Depths:
    for n in n_estimators:
        RF = RandomForestRegressor(max_depth= depth, n_estimators = n)
        RF = RF.fit(x_2_train, y_2_train)
        predition = RF.predict(x_2_test)
        predition_list = predition.tolist()
        y_2_test_list = y_2_test.tolist()
        res_mse_list.append(mse(predition_list, y_2_test_list))  
        res_rmse_list.append(rmse(predition_list, y_2_test_list))
```

### Question: Bias and deviation Rabdom Forest VS Decision Tree
#### Bias and deviation
    Bias: Bias is the difference between the average prediction of our model and the correct value which we are trying to predict.
    
    Variance: Variance is the variability of model prediction for a given data point or a value which tells us spread of our data. Model with high variance pays a lot of attention to training data and does not generalize on the data which it hasn’t seen before.
####  Bias and deviation Rabdom Forest VS Decision Tree
    An algorithm like Decision Tree has low bias but high variance, because it can easily change as small change in input variable. In general, it does not generalize the pattern well. It leads to overfitting.
    
    random forest use each tree's high variance models and aggregate them in order to reduce variance and thus enhance prediction accuracy. Both Bagging and Random Forests use Bootstrap sampling, and as described in "Elements of Statistical Learning", this increases bias in the single tree.


#### Conclusion:
     random forests are a strong modeling technique and much more robust than a single decision tree. They aggregate many decision trees to limit overfitting as well as error due to bias and therefore yield useful results.
     
     And yes we can see that our model is acting the same way theory says, Random Forest showed a stronger algorithm.
