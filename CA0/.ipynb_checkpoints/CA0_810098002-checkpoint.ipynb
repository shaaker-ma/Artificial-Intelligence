##  <font color='green'>Question 3:</font>
### Part 3-3:
    Here is 6 solutions for missing data:
    
    1- Do Nothing:
        well this is the most easy thing there is to do, and obviously we are not going to use it compare to the average
        replacement
        
    2- Imputation Using (Mean/Median) Values:
        Our method above.
        
        -Disadvantages:
            -Mean imputation does not preserve the relationships among variables.
            -Mean Imputation Leads to An Underestimate of Standard Errors
            
        -Advantages:
            -Super easy
    
    3- Imputation Using Most Frequent or Zero/Constant Values: 
        Most Frequent is another statistical strategy to impute missing values 
        
       - Benifits:
            -Works well with categorical features.
            
       - Consequences:
            -It does NOT factor the correlations between features
            - can introduce bias in the data.
    
    4-Imputation Using k-NN: -> Better than 3 above
        KNN or k nearest neighbours is used for simple classification, it uses feature similarity to predict the values
        of any new data points.This means that the new point is assigned a value based on how closely it resembles the points
        in the training set.
        
        - Benifits:
            -Much more accurate than the mean, median or most frequent imputation methods.

 ## <font color='green'>Question 6: Average with vectorization</font>

    Vectorization: We would get average age of the price of vehicles which use gas and diesel.

```python
import time
tic = time.time()

result = np.select([(dfFilled['fueltype'] == 0)],['target']) #Gas
result = np.where(result== 'target')
avgPriceGas = dfFilled.iloc[result[0],12].mean()

result = np.select([(dfFilled['fueltype'] == 1)],['target']) #Diesel
result = np.where(result== 'target')
avgPriceDiesel = dfFilled.iloc[result[0],12].mean()

toc = time.time()
print("for Vectorization: %f ms" %((toc-tic)*1000))
print("avgPriceGas: ", avgPriceGas)
print("avgPriceDiesel: ", avgPriceDiesel)
```

 ## <font color='green'>Question 7: Average with loop</font>

    Vectorization: We would get average age of the price of vehicles which use gas and diesel.

```python
tic = time.time()

sumPriceGas =0
numPriceGas =0
sumPriceDiesel =0
numPriceDiesel =0
for index, row in dfFilled.iterrows():
    if row['fueltype'] == 0 and ((index+1) not in carID_NANPrice):
        sumPriceGas = sumPriceGas + row['price']
        numPriceGas = numPriceGas+1
    elif row['fueltype'] == 1 and ((index+1) not in carID_NANPrice):
        sumPriceDiesel = sumPriceDiesel + row['price']
        numPriceDiesel = numPriceDiesel+1
toc = time.time()
print("for loop: %f ms" %((toc-tic)*1000))
print("avgPriceGas: ", sumPriceGas/numPriceGas)
print("avgPriceDiesel: ", sumPriceDiesel/numPriceDiesel)
```

We can totally see that vectorization helps us to make our code more efficient both in timing and also the number of lines that we use.

```python
point_1 = np.where(dfFilledNormal['enginesize'] >= 250)
print("first point desc:")
print(dfFilledNormal.iloc[47])
point_2 = np.where(dfFilledNormal['enginesize'] <= 63)
print("\n\nsecond point desc:")
print(dfFilledNormal.iloc[18])
```

```python
dfFilledNormal.iloc[18].price
```

```python
x0 = dfFilledNormal.iloc[18].enginesize
y0 = dfFilledNormal.iloc[18].price
x1 = dfFilledNormal.iloc[47].enginesize
y1 =  dfFilledNormal.iloc[47].price
m = (y1 - y0)/(x1 - x0)
print("Theta1: ",m)
```

```python
def hypothesis(engineSize):
    h = m*(engineSize - x0) + y0
    return h
```

Hypothesis test:

```python
print(hypothesis(200))
```

```python
import math
dfFilledNormalZippedPredicted = dfFilledNormalZipped.copy()
h_theta = dfFilledNormalZipped.copy()
for index, row in dfFilledNormalZippedPredicted.iterrows():
    if math.isnan(row['price']) == True:
        row['price'] = hypothesis(row['enginesize'])
for index, row in h_theta.iterrows():
    row['price'] = hypothesis(row['enginesize'])
```

##  <font color='green'>Question 12: Calculating MSE of prediced function</font>

```python
h_theta_list = h_theta['price'].tolist()
y = dfFilledNormalZippedPredicted['price'].tolist()
sumOfSquare = 0
for i in range(len(h_theta_list)):
    sumOfSquare = sumOfSquare + ((y[i] - h_theta_list[i])**2)
MSE = sumOfSquare/len(y)
print("Maximum Square Error: ", MSE)
print("Which is less than 0.5")
```

##  <font color='green'>Question 13: Sketching both the predicted function and the data we had earlie /font>

```python
import matplotlib.pyplot as plt

x = dfFilledNormalZippedPredicted['enginesize'].tolist()
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.scatter(x, y, s=10, c='b', marker="s", label='y')
ax1.scatter(x, h_theta_list, s=10, c='r', marker="o", label='h_theta')
plt.legend(loc='upper left');
plt.show()
```

We can totaly see that we have predicted a linear hypothesis which fits the input (engine size) to the output (price), in the
previous section we saw that the Maximum Square Error of this function is 0.3561399942967544 and is a very suitable function to be used
