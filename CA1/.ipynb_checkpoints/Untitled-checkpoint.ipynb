```python
n,m = [],[]
maze = []
startPoint_all = []
finalPoint_all = []
bagCapacity_all = []
numBalls_all = []
ballsPos_all = []
maze_all = [[],[],[],[]]
for i in range(4):
    file = open(str(i+1)+".txt",'r')
    lines_i = file.readlines()
    print(maze)
    maze_all[i].append(maze)
    maze.clear() 
    print(maze_all)
    print("\n\n\n")
    count = 0
    countBallsTemp = 0
    for line in lines_i:
        line = line.split()
        if(count ==0):
            n.append(int(line[0]))
            m.append(int(line[1]))
            count +=1
        elif(count ==1):
            startPoint = [int(line[0]),int(line[1])]
            startPoint_all.append(startPoint)
            count +=1
        elif(count == 2):
            finalPoint = [int(line[0]),int(line[1])]
            finalPoint_all.append(finalPoint)
            count +=1
        elif(count == 3):
            bagCapacity = int(line[0])
            bagCapacity_all.append(bagCapacity)
            count +=1
        elif(count == 4):
            numBalls = int(line[0])
            numBalls_all.append(numBalls)
            countBallsTemp = numBalls
            ballsPos = {}
            count +=1
            if(numBalls==0): ballsPos_all.append({})
        elif(count >4 and countBallsTemp>0):
            ballPos_i = {(int(line[0]),int(line[1])):(int(line[2]),int(line[3]))}
            ballsPos.update(ballPos_i)
            countBallsTemp -=1
            count +=1
            if(countBallsTemp == 0): #all balls are counted and "ballsPos" is complete
                ballsPos_all.append(ballsPos)
        else:
            maze.append(line)
print("Maze is a ", n, "*", m, "map")
print("Start point is: ", startPoint_all)
print("Final point is: ", finalPoint_all)
print("Bag capacity is:", bagCapacity_all)
print("Number of balls is:", numBalls_all)
print("ballsPos", ballsPos_all)

# # for i in range(len(maze_all)):
# #     print(" ",maze_all[i],"\n")
# #     print("\n\n")
```

```python
class Node(object):
    def __init__(self, _state):
        self.state = tuple(_state)
        self.parent = None
        self.depth = 0
    def printState(self):
        print(self.state)
    def setParent(self, _Node):
        self.parent = _Node
    def setDepth(self, _depth):
        self.depth = _depth
    def getState(self):
        return self.state
    def getParent(self):
        return self.parent
    def getDepth(self):
        return self.depth
    def printPath(self,Node):
        while(Node.parent != None):
            Node.printState()
            Node = Node.parent
        Node.printState()
```

```python
a = Node((0,0))
v = Node((0,1))
b = Node((0,2))
n = Node((0,3))
m = Node((0,4))
del a
v.setParent(a)
b.setParent(v)
n.setParent(b)
m.setParent(n)
m.printPath(m)
```
