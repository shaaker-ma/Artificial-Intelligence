```python
import random
def mathFunc(x,y,z):
    return (x+3)**2 + 8*y**2 + 23*z**2  
def fitness(x,y,z):
    valueOfMathFunc = mathFunc(x,y,z)
    if(valueOfMathFunc == 0):
        return 99999
    else:
        return (1/valueOfMathFunc)
class Chromosome:
    def __init__(self, x, y, z):
        self.position = (x, y, z)
        self.rate = fitness(x, y, z)
    def getPosition(self):
        return self.position
    def getRate(self):
        return self.rate
    
def geneticAlgorithm():
    firstPopulation = []
    for i in range(1000): #1000 chromosome
        newChromosome = Chromosome( random.uniform(0,1000),
                                    random.uniform(0,1000),
                                    random.uniform(0,1000))
        firstPopulation.append([newChromosome.getRate(),newChromosome])
    firstPopulation.sort()
    firstPopulation.reverse()
    rankedPopulation = firstPopulation
    while(rankedPopulation[0][0]<1000):
        
        finalNewPopulation = []
        
        bestPopulation = rankedPopulation[:100]
        bestGenes = [[0 for x in range(0)] for y in range(3)] 
        for ind in bestPopulation:
            bestGenes[0].append((ind[1].getPosition())[0])
            bestGenes[1].append((ind[1].getPosition())[1])
            bestGenes[2].append((ind[1].getPosition())[2])
        for i in range(len(bestPopulation)):
            x = random.choice(bestGenes[0])
            y = random.choice(bestGenes[1])
            z = random.choice(bestGenes[2])
            newChromosome = Chromosome(x, y, z)
            finalNewPopulation.append([newChromosome.getRate(),newChromosome])
#         print(finalNewPopulation)
        restPopulation = rankedPopulation[100:]
        restGenes = [[0 for x in range(0)] for y in range(3)] 
        for ind in restPopulation:
            restGenes[0].append((ind[1].getPosition())[0])
            restGenes[1].append((ind[1].getPosition())[1])
            restGenes[2].append((ind[1].getPosition())[2])
        for i in range(len(restGenes)):
            x = random.choice(restGenes[0])
            y = random.choice(restGenes[1])
            z = random.choice(restGenes[2])
            newChromosome = Chromosome(x, y, z)
            finalNewPopulation.append([newChromosome.getRate(),newChromosome])
#         sorted(finalNewPopulation, key=operator.itemgetter(0))
        sorted(finalNewPopulation, key=lambda x: x[0])
        finalNewPopulation.reverse()
        rankedPopulation = finalNewPopulation
    print('done')
geneticAlgorithm()
```
